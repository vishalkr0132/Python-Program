# 2. Question: Write a function to compute the factorial of a number using recursion.

```python
def factorial(numbers):
    if numbers == 0 or numbers == 1:
        return 1
    else:
        return numbers * factorial(numbers - 1)

number = int(input('Enter a number to find its factorial: '))
result = factorial(number)
print(f"The factorial of {number} is: {result}")
```

# 3. Question: You are given a list of strings. Write a function to filter out all strings that are palindromes.

```python
def is_palindrome(string):
    length = len(string)
    length_str = length // 2
    for i in range(length_str):
        if string[i] != string[length - i - 1]:
            return False
    return True

def filter_palindromes(strings):
    lst = []
    for string in strings:
        if is_palindrome(string):
            lst.append(string)
    return lst

strings = ["radar", "hello", "level", "world", "python", "madam"]
print(filter_palindromes(strings))
```

# 4. Question: Given a string, write a function to check if it is an anagram of another string.

```python
def string_length(string):
  count = 0
  for char in string:
    count += 1
  return count
def anagram(string1, string2):
  length1 = string_length(string1)
  length2 = string_length(string2)

  if length1 != length2:
    return False
  count1 = {}
  count2 = {}
  for char in string1:
    if char in count1:
      count1[char] += 1
    else:
      count1[char] = 1

  for char in string2:
    if char in count2:
      count2[char] += 1
    else:
      count2[char] = 1

  for char in count1:
    if count1[char] != count2.get(char,0):
      return False
  return True

string1 = input('Enter the first string: ')
string2 = input('Enter the second string: ')
print(anagram(string1, string2))
```

# 6. Question: Given two lists, write a function that returns the elements that are common to both lists.

```python
def common(list1, list2):
  common_elements = []
  for element1 in list1:
    for element2 in list2:
      if element1 == element2:
        if element1 not in common_elements:
          common_elements.append(element1)
        break
  return common_elements

list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
print(common(list1, list2))
```

# 8. Question: Write a Python function to merge two dictionaries. If both dictionaries have the same key, prefer the second dictionary's value.

```python
def merge_dict(dict1, dict2):
  merge_dict= {}
  for key in dict1:
    merge_dict[key] = dict1[key]
  for key in dict2:
    merge_dict[key] = dict2[key]
  return merge_dict

dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 20, 'c': 30, 'd': 40}
merged_dict = merge_dict(dict1, dict2)
print("Merged dictionary:", merged_dict)
```

# 10. Question: Write a function that checks if a string contains all letters of the alphabet at least once.

```python
def contains_all_alphabet(string):
  alphabet = {}
  for char in string:
    if 'A' <= char <= 'Z':
      index = chr(ord(char) + (ord('a') - ord('A')))
      alphabet[index] = True
    elif 'a' <= char <= 'z':
      alphabet[char] = True

  if len(alphabet) == 26:
    return True
  else:
    return False

string = input('Enter a string: ')
print(contains_all_alphabet(string))
```

# 11. Question: Write a function that checks if a given string is a valid IPv4 address.

```python
def is_valid_ipv4(ip):
    # Initialize a list to store parts of the IP address
    parts = ["", "", "", ""]
    part_index = 0

    # Iterate over each character in the string
    for char in ip:
        if char == '.':
            # Move to the next part
            part_index += 1
            if part_index > 3:  # More than 3 dots means it's invalid
                return False
        else:
            # Append the character to the current part
            if part_index < 4:
                parts[part_index] += char
            else:
                return False  # More than 4 parts is invalid

    # Check if we have exactly 4 parts
    if part_index != 3:
        return False

    # Validate each part
    for part in parts:
        # Check if the part is not empty and only contains digits
        if part == "":
            return False

        # Convert the part to an integer manually
        num = 0
        for digit in part:
            if digit < '0' or digit > '9':  # Ensure it's a digit
                return False
            num = num * 10 + (ord(digit) - ord('0'))

        # Check if the number is within the valid range
        if num < 0 or num > 255:
            return False
        if len(part) > 1 and part[0] == '0':
            return False
    return True
address = "192.168.1.1"
print(is_valid_ipv4(address))
```

12. Question: Given a list of numbers, write a function to compute the mean, median, and mode.

```python
def mean(numbers):
  total_sum = 0
  count = 0
  for num in numbers:
    total_sum += num
    count += 1
    mean = total_sum / count
    return mean

def median(numbers):
  n = 0
  for i in numbers:
    n += 1
  for j in range(n):
    for k in range(0,n-j-1):
      if numbers[k] > numbers[k + 1]:
        numbers[k], numbers[j + 1] = numbers[k + 1], numbers[k]
  if n % 2 == 1:
    median = numbers[n // 2]
  else:
    median = (numbers[n // 2 - 1] + numbers[n // 2]) / 2
  return median

def mode(numbers):
  frequency = {}
  for num in numbers:
    if num in frequency:
      frequency[num] += 1
    else:
      frequency[num] = 1
  max_count = 0
  mode = None
  for num in frequency:
    if frequency[num] > max_count:
      max_count = frequency[num]
      mode = num
  return mode

numbers = [4, 1, 2, 2, 3, 5, 4, 2, 6]

mean = mean(numbers)
median = median(numbers)
mode = mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

# 13. Question: Write a function to compute the Fibonacci series up to n.

```python
def fibonacci_series(numbers):
  fib_series = [0] * numbers # Allocate space for numbers of elements
  if numbers > 0:
    fib_series[0] = 0
  if numbers > 1:
    fib_series[1] = 1
  current_index = 2
  while True:
    next_fib = fib_series[current_index - 1] + fib_series[current_index - 2]
    if next_fib >= numbers:
      break
    fib_series[current_index] = next_fib
    current_index += 1
  return fib_series[:current_index]

numbers = int(input('Enter a number: '))
print(fibonacci_series(numbers))
```

# 17. Question: Write a function that returns the longest consecutive subsequence in a list of numbers.

```python
def lcs(num):
    if not num:
        return None

    num_unique = []
    for i in num:
        if i not in num_unique:
            num_unique += [i]

    num_len = 0
    for j in num_unique:
        num_len += 1

    count_num_unique = 0
    for count in num_unique:
      count_num_unique += 1

    for k in range(num_len):
        for l in range(0, num_len - k - 1):
            if num_unique[l] > num_unique[l + 1]:
                num_unique[l], num_unique[l + 1] = num_unique[l + 1], num_unique[l]
    prev = []
    curr = [num_unique[0]]

    for m in range(1, count_num_unique):
        if num_unique[m] - num_unique[m - 1] == 1:
            curr += [num_unique[m]]
        else:
            if len(curr) > len(prev):
                prev = curr
            curr = [num_unique[m]]

    return prev if len(prev) > len(curr) else curr

num = [100, 4, 200, 1, 3, 2]
print(lcs(num))
```

# 21. Question: Write a function that checks if a given binary tree is a valid binary search tree.

```python
class TreeNode:
  def __init__(self,value=0,left = None,right = None):
    self.value = value
    self.left = left
    self.right = right

def is_valid_bst(root, left= None,right=None):
  if not root:
    return True
  if left and root.value <= left.value:
    return False
  if right and root.value >= right.value:
    return False
  return is_valid_bst(root.left,left,root) and is_valid_bst(root.right,root,right)
root = TreeNode(2,TreeNode(1), TreeNode(3))
print(is_valid_bst(root))
```

# 23. Question: Write a function that returns the intersection of two sorted arrays. Assume each array does not have duplicates.

```python
def intersection_of_sorted_arrays(nums1, nums2):
  i,j = 0,0
  intersection = []
  while i < len(nums1) and j < len(nums2):
    if nums1[i] == nums2[j]:
      intersection.append(nums1[i])
      i += 1
      j += 1
    elif nums1[i] < nums2[j]:
      i +=1
    else:
        j += 1
  return intersection

arr1 = [1,2,3,4,5,6]
arr2 = [2,3,5,7]
print(intersection_of_sorted_arrays(arr1,arr2))
```
